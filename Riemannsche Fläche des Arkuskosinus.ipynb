### Visualisierung $k$-tes Blatt der Riemannschen Fläche des Arkuskosinus Arccos

```python
#Farbkodierung Blatt B_{k}
u = np.linspace(0, np.pi, 50)
v = np.linspace(-3, 3, 50)
uGrid, vGrid = np.meshgrid(u, v)

blatt_k = go.Surface(
x = np.cos(uGrid)*np.cosh(vGrid),
y = -np.sin(uGrid)*np.sinh(vGrid),
z = uGrid,
showscale=True,
colorbar = dict(
            title = 'Re(Arccos(z))',
            titleside = 'top',
            tickmode = 'array',
            ticks = 'outside',
            tickvals = [0,np.pi/4,np.pi/2, 3*np.pi/4, np.pi],
            ticktext = ['0+2kπ','π/4+2kπ','π/2+2kπ', '3/4π +2kπ', 'π+2kπ'],
            len = 0.5),
colorscale='Viridis',
reversescale = False
)

data = [blatt_k]

layout = go.Layout(
                    title='Visualisierung k-tes Blatt',
                    scene = dict(
                    xaxis = dict(
                       title='Re(z)',
                    range=[-5, 5],
                    ),
                    yaxis = dict(
                        title='Im(z)',
                    range=[-5, 5],
                    ),
                    zaxis = dict(
                        title='Re(Arccos(z))', showticklabels=False),),
                    width=900,
                    height=900,
                  )

fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='Visualisierung k-tes Blatt Arccos')
```
