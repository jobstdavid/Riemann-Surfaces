### Visualisierung $k$-tes Blatt der Riemannschen Fläche des Arkustangens Arctan

```python
#Farbkodierung Blatt B_{k}
u = np.linspace(-np.pi/2+0.01, np.pi/2-0.01, 50)
v = np.linspace(-2, 2, 50)
uGrid, vGrid = np.meshgrid(u, v)

blatt_k = go.Surface(
x = (np.sin(uGrid)*np.cos(uGrid))/(np.cos(uGrid)**2 + np.sinh(vGrid)**2),
y = (np.sinh(vGrid)*np.cosh(vGrid))/(np.cos(uGrid)**2 + np.sinh(vGrid)**2),
z =  uGrid,
showscale=True,
colorbar = dict(
            title = 'Re(Arctan(z))',
            titleside = 'top',
            tickmode = 'array',
            ticks = 'outside',
            tickvals = [-np.pi/2+0.01,-np.pi/4,0, np.pi/4, np.pi/2-0.01],
            ticktext = ['-π/2+kπ','-π/4+kπ','0+kπ', 'π/4+kπ', 'π/2+kπ'],
            len = 0.5),
colorscale='Viridis',
)

data = [blatt_k]

layout = go.Layout(
                    title='Visualisierung k-tes Blatt',
                    scene = dict(
                    xaxis = dict(
                       title='Re(z)',
                    range=[-5, 5],
                    ),
                    yaxis = dict(
                        title='Im(z)',
                    range=[-5, 5],
                    ),
                    zaxis = dict(
                        title='Re(Arctan(z))', showticklabels=False),),
                    width=900,
                    height=900,
                  )

fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='Visualisierung k-tes Blatt Arctan')
```

### Visualisierung der Riemannschen Fläche des Arkustangens Arctan

```python
#Farbkodierung Blatt B_{-1}
u = np.linspace(-3*np.pi/2+0.005, -1*np.pi/2-0.005, 50)
v = np.linspace(-2, 2, 50)
uGrid, vGrid = np.meshgrid(u, v)

blatt_minus_1 = go.Surface(
x = (np.sin(uGrid)*np.cos(uGrid))/(np.cos(uGrid)**2 + np.sinh(vGrid)**2),
y = (np.sinh(vGrid)*np.cosh(vGrid))/(np.cos(uGrid)**2 + np.sinh(vGrid)**2),
z =  uGrid,
showscale=False,
colorscale='Viridis',
)

#Farbkodierung Blatt B_{0}
u = np.linspace(-1*np.pi/2+0.005, 1*np.pi/2-0.005, 50)
v = np.linspace(-2, 2, 50)
uGrid, vGrid = np.meshgrid(u, v)

blatt_0 = go.Surface(
x = (np.sin(uGrid)*np.cos(uGrid))/(np.cos(uGrid)**2 + np.sinh(vGrid)**2),
y = (np.sinh(vGrid)*np.cosh(vGrid))/(np.cos(uGrid)**2 + np.sinh(vGrid)**2),
z =  uGrid,
showscale=False,
colorscale='Viridis',
)

#Farbkodierung Blatt B_{1}
u = np.linspace(1*np.pi/2+0.005, 3*np.pi/2-0.005, 50)
v = np.linspace(-2, 2, 50)
uGrid, vGrid = np.meshgrid(u, v)

blatt_1 = go.Surface(
x = (np.sin(uGrid)*np.cos(uGrid))/(np.cos(uGrid)**2 + np.sinh(vGrid)**2),
y = (np.sinh(vGrid)*np.cosh(vGrid))/(np.cos(uGrid)**2 + np.sinh(vGrid)**2),
z =  uGrid,
showscale=False,
colorscale='Viridis',
)

#Verknüpfung der einzelnen Blätter
data = [blatt_minus_1, blatt_0, blatt_1]

layout = go.Layout(
                    title='Riemmansche Fläche des Arkustangens Arctan',
                    scene = dict(
                    xaxis = dict(
                       title='Re(z)',
                    range=[-2, 2],
                    ),
                    yaxis = dict(
                        title='Im(z)',
                    range=[-2, 2],
                    ),
                    zaxis = dict(
                        title='Re(Arctan(z))',),),
                    width=1000,
                    height=1000,
                  )

fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='Riemannsche Fläche des Arkustangens')
```
