### Visualisierung $0$-tes Blatt der Riemannschen Fläche der Quadratwurzel $q^{-1}$

```python
#Farbkodierung Blatt B_{0}
s = np.linspace(0, 1.0, 240)
t = np.linspace(-np.pi, np.pi, 240)
tGrid, sGrid = np.meshgrid(t, s)
rho = sGrid

x=rho*np.cos(tGrid)
y=rho*np.sin(tGrid)
z = x + 1j * y
r = np.absolute(z)
argument = np.angle(z)
wmod = r**(1./2)
argument /= 2
Imw = wmod * np.sin(argument)

blatt_0 = go.Surface(
x = r * np.cos(tGrid),
y = r * np.sin(tGrid),
z =  wmod * np.sin(argument),
showscale=True, 
colorbar = dict(
            title = 'Re(q(z))',
            titleside = 'top',
            tickmode = 'array',
            ticks = 'outside',
            len = 0.5),
colorscale='Viridis',
)

data = [blatt_0]

layout = go.Layout(
                    title='Visualisierung 0-tes Blatt',
                    scene = dict(
                    xaxis = dict(
                       title='Re(z)',),
                    yaxis = dict(
                        title='Im(z)',),
                    zaxis = dict(
                        title='Re(q(z))', showticklabels=False),),
                    width=800,
                    height=800,
                  )

fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='Visualisierung 0-tes Blatt')
```

### Visualisierung der Riemannschen Fläche der Quadratwurzel $q^{-1}$

```python
s = np.linspace(0, 1.0, 240)
t = np.linspace(-np.pi, np.pi, 240)
tGrid, sGrid = np.meshgrid(t, s)
rho = sGrid

x=rho*np.cos(tGrid)
y=rho*np.sin(tGrid)
z = x + 1j * y
r = np.absolute(z)
argument = np.angle(z)
wmod = r**(1./2)
argument /= 2
Imw = wmod * np.sin(argument)

trace0 = go.Surface(
x = r * np.cos(tGrid),
y = r * np.sin(tGrid),
z =  wmod * np.sin(argument),
showscale=False,
colorscale='Viridis',
)

trace1 = go.Surface(
x = r * np.cos(tGrid),
y = r * np.sin(tGrid),
z =  wmod * np.sin(argument+np.pi),
showscale=False,
colorscale='Viridis',
)

data = [trace0,trace1]

layout = go.Layout(
                    title='Riemmansche Fläche der Quadratwurzel',
                    scene = dict(
                    xaxis = dict(
                       title='Re(z)',),
                    yaxis = dict(
                        title='Im(z)',),
                    zaxis = dict(
                        title='Re(q(z))',),),
                    width=800,
                    height=800,
                  )

fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='Riemannsche Fläche der Quadratwurzel')
```
