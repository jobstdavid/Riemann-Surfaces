### Visualisierung $k$-tes Blatt der Riemannschen Fläche des Logarithmus Log

```python
#Farbkodierung Blatt B_{k}
s = np.linspace(0, 3, 240)
t = np.linspace(-1*np.pi, 1*np.pi - np.pi/10000, 240)
tGrid, sGrid = np.meshgrid(t, s)
r = sGrid

blatt_k = go.Surface(
x = r * np.cos(tGrid),
y = r * np.sin(tGrid),
z = np.arctan2(r*np.sin(tGrid),r*np.cos(tGrid)),
showscale=True,
colorbar = dict(
            title = 'Im(Log(z))',
            titleside = 'top',
            tickmode = 'array',
            ticks = 'outside',
            tickvals = [-np.pi,-np.pi/2,0, np.pi/2, np.pi],
            ticktext = ['-π+2kπ','-π/2+2kπ','0+2kπ', 'π/2+2kπ', 'π+2kπ'],
            len = 0.5),
colorscale='Viridis'
)

data = [blatt_k]

layout = go.Layout(
                    title='Visualisierung k-tes Blatt',
                    scene = dict(
                    xaxis = dict(
                       title='Re(z)',),
                    yaxis = dict(
                        title='Im(z)',),
                    zaxis = dict(
                        title='Im(Log(z))', showticklabels=False),),
                    width=800,
                    height=800,
                  )

fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='Visualisierung k-tes Blatt Log')
```

### Visualisierung der Riemannschen Fläche des Logarithmus Log

```python
s = np.linspace(0, 1, 240)
t = np.linspace(-1*np.pi, 1*np.pi - np.pi/10000, 240)
tGrid, sGrid = np.meshgrid(t, s)
r = sGrid

#Farbkodierung Blatt B_{-1}
blatt_minus_1 = go.Surface(
x = r * np.cos(tGrid),
y = r * np.sin(tGrid),
z = np.arctan2(r*np.sin(tGrid),r*np.cos(tGrid))-2*1*np.pi, 
showscale=False,
colorscale='Viridis',
)

#Farbkodierung Blatt B_{0}
blatt_0 = go.Surface(
x = r * np.cos(tGrid),
y = r * np.sin(tGrid),
z = np.arctan2(r*np.sin(tGrid),r*np.cos(tGrid)),
showscale=False,
colorscale='Viridis',
)

#Farbkodierung Blatt B_{1}
blatt_1 = go.Surface(
x = r * np.cos(tGrid),
y = r * np.sin(tGrid),
z = np.arctan2(r*np.sin(tGrid),r*np.cos(tGrid))+2*1*np.pi,  
showscale=False,
colorscale='Viridis',
)

#Verknüpfung der einzelnen Blätter
data = [blatt_minus_1, blatt_0, blatt_1]

layout = go.Layout(
                    title='Riemmansche Fläche des Logarithmus Log',
                    scene = dict(
                    xaxis = dict(
                       title='Re(z)',),
                    yaxis = dict(
                        title='Im(z)',),
                    zaxis = dict(
                        title='Im(Log(z))',),),
                    width=800,
                    height=800,
                  )

fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='Riemannsche Fläche des Logarithmus')
```
