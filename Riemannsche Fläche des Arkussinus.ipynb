### Visualisierung $k$-tes Blatt der Riemannschen Fläche des Arkussinus Arcsin

```python
#Farbkodierung Blatt B_{k}
u = np.linspace(-np.pi/2, np.pi/2, 50)
v = np.linspace(-3, 3, 50)
uGrid, vGrid = np.meshgrid(u, v)

blatt_k = go.Surface(
x = np.sin(uGrid)*np.cosh(vGrid),
y = np.cos(uGrid)*np.sinh(vGrid),
z =  uGrid,
showscale=True,
colorbar = dict(
            title = 'Re(Arcsin(z))',
            titleside = 'top',
            tickmode = 'array',
            ticks = 'outside',
            tickvals = [-np.pi/2,-np.pi/4,0, np.pi/4, np.pi/2],
            ticktext = ['-π/2+2kπ','-π/4+2kπ','0+2kπ', 'π/4+2kπ', 'π/2+2kπ'],
            len = 0.5),
colorscale='Viridis',
)

data = [blatt_k]

layout = go.Layout(
                    title='Visualisierung k-tes Blatt', 
                    scene = dict(
                    xaxis = dict(
                       title='Re(z)',
                    range=[-5, 5],
                    ),
                    yaxis = dict(
                        title='Im(z)',
                    range=[-5, 5],
                    ),
                    zaxis = dict(
                        title='Re(Arcsin(z))', showticklabels=False),),
                    width=900,
                    height=900,
                  )

fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='Visualisierung k-tes Blatt Arcsin')
```
